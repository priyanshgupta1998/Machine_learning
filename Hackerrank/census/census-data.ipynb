```
import pandas as pd
import numpy as np
columns = ['age' , 'sex' , 'education' , 'country' ,'race' , 'status', 'workclass','occupation', 'hours-per-week' , 'income' , 'capital-gain' , 'capital-loss']
data = pd.read_csv('/census.csv' ,names=columns )
print(data.shape )
cols = list(data.columns)[:10]
for i in range(len(cols)):
  dfDummies = pd.get_dummies(data[cols[i]])
  data = pd.concat([data, dfDummies], axis=1)
data.drop(( list(data.columns)[:10]) ,axis = 1 ,inplace =True)
dum_cpgain = pd.get_dummies(data['capital-gain']) 
dum_cploss = pd.get_dummies(data['capital-loss'])
data = pd.concat([data,dum_cpgain ,dum_cploss  ], axis=1)
data.drop(( ['capital-loss=High','capital-gain=High','capital-gain=Low','capital-loss=Low' ]) ,axis = 1 ,inplace =True)
print(data.shape)
```
